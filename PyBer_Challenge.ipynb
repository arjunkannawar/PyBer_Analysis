## Deliverable 1: Get a Summary DataFrame 

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Create the Suburban, Rural and urban city DataFrames.
suburban_cities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]
rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]
urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]

#  1. Get the total rides for each city type
type_rides_count = pyber_data_df.groupby(["type"]).count()["ride_id"]
type_rides_count
```

```python
# 2. Get the total drivers for each city type
# In Step 2, use the groupby() function to create a Series of data that has the 
# type of city as the index, then apply the sum() method to the "driver_count" column.
type_drivers_count = city_data_df.groupby(["type"]).sum()["driver_count"]
type_drivers_count
```

```python
#  3. Get the total amount of fares for each city type
type_fare_count = pyber_data_df.groupby(["type"]).sum()["fare"]
type_fare_count
```

```python
#  4. Get the average fare per ride for each city type. 
#In Step 4, calculate the average fare per ride by city type by 
#dividing the sum of all the fares by the total rides.

city_avg_fare = pyber_data_df.groupby(["type"]).mean()["fare"]
city_avg_fare.head()
```

```python
# 5. Get the average fare per driver for each city type. 
#In Step 5, calculate the average fare per driver by city type by dividing the 
# sum of all the fares by the total drivers.
type_avg_fare = type_fare_count / type_drivers_count
type_avg_fare.head()
```

```python
#  6. Create a PyBer summary DataFrame. 
# df = pd.concat([a_series, another_series], axis=1)
Pyber_summary_df = pd.concat([type_rides_count, type_drivers_count, type_fare_count, city_avg_fare, type_avg_fare], axis =1)
Pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
Pyber_summary_df.index.name = None
Pyber_summary_df
```

```python
#  8. Format the columns.
Pyber_summary_2_df = Pyber_summary_df.set_axis(['Total Rides', 'Total Drivers', 'Total Fares', 'Average fare Per Ride', 'Average Fare per driver'], axis=1, inplace=False)
Pyber_summary_2_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
# Combine the data into a single dataset
pyber_data_merged_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])
pyber_data_merged_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
# type_drivers_count = pyber_data_df.groupby(["type"]).sum()["driver_count"]
pyber_data_new_df = pyber_data_merged_df.groupby(["type" , "date"]).sum()[["fare"]]
pyber_data_new_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
pyber_data_merged_df = pyber_data_merged_df.reset_index()
pyber_data_merged_df.tail()
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
#df.pivot(index='foo', columns='bar', values='baz')
pyber_data_merged_df = pyber_data_merged_df.pivot(index ='date', columns ='type', values ='fare')
pyber_data_merged_df
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
pyber_date_loc_df = pyber_data_merged_df.loc['2019-01-01':'2019-04-29']
pyber_date_loc_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
pyber_date_loc_df.index = pd.to_datetime(pyber_date_loc_df.index)
pyber_date_loc_df
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
pyber_date_loc_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
sum_fares_weekly = pyber_date_loc_df.resample("W").sum()
sum_fares_weekly
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')


#Create the plot with ax.plt()
sum_fares_weekly.plot()
plt.ylabel('Fare ($USD)')
plt.xlabel('')
plt.title("Total Fare By City type")
#plt.figure(figsize=(10, 6))
plt.savefig("analysis/PyBer_fare_summary.png")
```
